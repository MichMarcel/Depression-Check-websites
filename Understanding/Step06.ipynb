## Step 6 — Basic Quality Scan

```python
import pandas as pd

# Load dataset (ubah path sesuai lokasi file Anda)
file_path = r"C:\Users\Michael Marcel\OneDrive\Desktop\TA\Data\Dataset\student_lifestyle_100k.csv"

df = pd.read_csv(file_path)

df.head()
```

```python
# Step 6.1 — Missing Values per Column
missing_values = df.isnull().sum()

print('Jumlah Missing Values per Kolom:')
missing_values
```

```python
# Persentase Missing Values
missing_percent = (df.isnull().sum() / len(df)) * 100

missing_percent
```

```python
# Unique Value Count per Column
unique_counts = df.nunique()

print('Jumlah Nilai Unik per Kolom:')
unique_counts
```

```python
# Unique values khusus kolom kategorikal
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    print(f'\n{col} : {df[col].nunique()} unique values')
    print(df[col].value_counts().head())
```

```python
# Step 6.3 — Range Nilai Numerik
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

for col in numeric_cols:
    print(f'\n{col}')
    print('Min :', df[col].min())
    print('Max :', df[col].max())
```

```python
import os
import zipfile
import pandas as pd
from kaggle.api.kaggle_api_extended import KaggleApi

# Authenticate with Kaggle API
# Requires ~/.kaggle/kaggle.json with your API credentials
api = KaggleApi()
api.authenticate()

# Define dataset details
DATASET = "aldinwhyudii/student-depression-and-lifestyle-100k-data"
DOWNLOAD_DIR = "data"

# Download the dataset
os.makedirs(DOWNLOAD_DIR, exist_ok=True)
api.dataset_download_files(DATASET, path=DOWNLOAD_DIR, unzip=True)

print(f"Dataset downloaded to '{DOWNLOAD_DIR}/'")
print("Files:", os.listdir(DOWNLOAD_DIR))
```

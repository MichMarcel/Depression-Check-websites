# Kategorisasi Variabel Numerik

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
file_path = "student_lifestyle_100k.csv"
df = pd.read_csv(r"C:\Users\Michael Marcel\OneDrive\Desktop\TA\Data\Dataset\student_lifestyle_100k.csv")
```

## Binning per Variabel

- Sleep Duration

```python
bins_sleep = [-0.01, 7, 9, 24]
labels_sleep = ['Rendah', 'Normal', 'Tinggi']

df['Sleep_Category'] = pd.cut(df['Sleep_Duration'], bins=bins_sleep, labels=labels_sleep)
```

```python
df['Sleep_Category'].value_counts().reset_index()
```

```python
sleep_table = df['Sleep_Category'].value_counts().reset_index()
sleep_table.columns = ['Kategori', 'Jumlah']

plt.figure()
plt.pie(sleep_table['Jumlah'], labels=sleep_table['Kategori'], autopct='%1.1f%%')
plt.title('Distribusi Sleep Category')
plt.show()
```

- Study Hours

```python
bins_study = [-0.01, 1, 2, 24]
labels_study = ['Rendah', 'Normal', 'Tinggi']

df['Study_Category'] = pd.cut(df['Study_Hours'], bins=bins_study, labels=labels_study)
```

```python
df['Study_Category'].value_counts().reset_index()
```

```python
study_table = df['Study_Category'].value_counts().reset_index()
study_table.columns = ['Kategori', 'Jumlah']

plt.figure()
plt.pie(study_table['Jumlah'], labels=study_table['Kategori'], autopct='%1.1f%%')
plt.title('Distribusi Study Category')
plt.show()
```

## Binning by Depression

- Depression by Sleep Category

```python
pd.crosstab(df['Sleep_Category'], df['Depression'])
```

```python
pd.crosstab(df['Sleep_Category'], df['Depression'], normalize='index') * 100
```

```python
cats = ['Rendah', 'Normal', 'Tinggi']
labels = df['Depression'].unique()

plt.figure(figsize=(12,4))

for i, c in enumerate(cats):
    plt.subplot(1, len(cats), i+1)
    wedges, _, _ = plt.pie(df[df['Sleep_Category']==c]['Depression'].value_counts().reindex(labels, fill_value=0), autopct='%1.1f%%')
    plt.title(c)

plt.legend(wedges, labels, title="Depression", loc="lower center", bbox_to_anchor=(0.5, -0.2))
plt.tight_layout()
plt.show()
```

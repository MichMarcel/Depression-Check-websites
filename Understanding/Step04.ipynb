### Dataset Context / Purpose

Data berisi identitas dasar seperti usia, jenis kelamin, dan departemen, serta indikator performa akademik yang ditunjukkan melalui nilai Indeks Prestasi Kumulatif (CGPA). Dataset juga merekam berbagai aspek perilaku keseharian mahasiswa yang berpotensi memengaruhi kesejahteraan dan kinerja akademik.

Variabel gaya hidup yang diamati meliputi durasi tidur, lama waktu belajar, intensitas penggunaan media sosial yang berhubungan dengan keberhasilan akademik (IPK/CGPA) dan masalah kesehatan mental (Depression & Stress). Dataset ini juga memasukkan variabel psikologis seperti tingkat stres dan indikasi depresi, sehingga memungkinkan analisis tentang hubungan antara pola hidup, kesehatan mental, dan pencapaian akademik mahasiswa.

Secara umum, dataset ini bertujuan untuk mengidentifikasi faktor-faktor yang berpengaruh terhadap kesejahteraan dan performa akademik mahasiswa. Hasil analisis diharapkan dapat menjadi dasar pengambilan keputusan dalam upaya peningkatan kualitas hidup dan keberhasilan studi mahasiswa.

## Step 3 — Check Data Types

```python
import pandas as pd
```

```python
# Load dataset
file_path = "student_lifestyle_100k.csv"
df = pd.read_csv(r"C:\Users\Michael Marcel\OneDrive\Desktop\TA\Data\Dataset\student_lifestyle_100k.csv")

print("INFORMASI STRUKTUR DATASET")
df.info()
```

```python
# Menampilkan tipe data dalam bentuk tabel
print("TIPE DATA TIAP KOLOM")
dtype_table = pd.DataFrame({
    "Column Name": df.columns,
    "Data Type": df.dtypes.values
})

display(dtype_table)
```

```python
# Mengelompokkan tipe data
print("KELOMPOK TIPE DATA")

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
datetime_cols = df.select_dtypes(include=['datetime64']).columns.tolist()

print("Kolom Numerik:")
print(numeric_cols)

print("\nKolom Kategorikal:")
print(categorical_cols)

print("\nKolom Date/Time:")
print(datetime_cols)
```
